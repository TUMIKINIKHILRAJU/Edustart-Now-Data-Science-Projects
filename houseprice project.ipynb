```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python

data = pd.read_csv("C://Users//Bhargav//Desktop//kc_house_data.csv")
print (data)
```

```python
X = data.iloc[:, 3:].values
y = data.iloc[:, 2].values
print (X)
print (y)
```

```python
#Splitting the dataset into training a testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
regressor_LR = LinearRegression()
```

```python
regressor_LR.fit(X_train,y_train)
```

```python
y_pred_lin = regressor_LR.predict(X_test)
print (y_pred_lin)
```

```python
plt.plot(y_test)
plt.plot(y_pred_lin)
plt.xlabel('count')
plt.ylabel('TRUE AND PREDICTED VALUES ')
plt.show()
```

```python
submission=pd.DataFrame(data = {'price':y_pred_lin})

submission.to_csv('C://Users//Bhargav//Desktop//Sample.csv',index = False)
```
